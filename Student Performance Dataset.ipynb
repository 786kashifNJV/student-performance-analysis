```python
avg = student_df.groupby("studytime")["G3"].mean()

plt.bar(avg.index, avg.values)
plt.xlabel("Study Time Level")
plt.ylabel("Average Final Grade")
plt.title("Study Time vs Final Grade")
plt.show()
```

```python
import matplotlib.pyplot as plt

avg = student_df.groupby("studytime")["G3"].mean()

plt.bar(avg.index, avg.values)
plt.xlabel("Study Time Level (1=Low, 4=High)")
plt.ylabel("Average Final Grade")
plt.title("Effect of Study Time on Final Grades")

#Add numbers on bars
for i, v in enumerate(avg.values):
    plt.text(avg.index[i], v+0.05, round(v,2), ha='center')

plt.show()
```
