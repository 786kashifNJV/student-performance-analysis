# Student Performance Analysis

## Research Question
Does study time affect students’ final grades?

## Dataset Source
This dataset was taken from Kaggle and contains information about students’ academic performance and study habits.

Source:https://www.kaggle.com/
## Objective
The objective of this project is to analyze whether study time has an effect on students' final grades.


```python
# Importing required libraries for Data Analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the Dataset

## In this section, we load the student performance dataset using Pandas and display it to understand its structure.

# Load the dataset into DataFrame
student_df= pd.read_csv('Student Performance Dataset.csv')

# Display the dataset
student_df
```

```python
# Calculate average final grade for each study time level
student_df.groupby("studytime")["G3"].mean()
```

## Visualization of Study Time and Grades

The following graphs visualize the relationship between students' study time and their final grades. Different graph types are used to better understand trends and patterns in the data.


## Study Time vs Final Grade

This graph shows the relationship between students’ study time and their average final grades.

We want to see whether students who study more tend to perform better.


```python
# Calculate average final grades by study time
avg = student_df.groupby("studytime")["G3"].mean()

# Create bar chart
plt.figure(figsize=(8,5))

plt.bar(avg.index, avg.values, color='skyblue')

plt.xlabel("Study Time Level (1 = Low, 4 = High)")
plt.ylabel("Average Final Grade")
plt.title("Study Time vs Final Grade")

plt.grid(axis='y')

plt.show()
```

```python
# Average grade by study time
avg = student_df.groupby("studytime")["G3"].mean()

plt.figure(figsize=(8,5))

plt.bar(
    avg.index,
    avg.values,
    color=['skyblue', 'lightgreen', 'orange', 'salmon']
)

plt.xlabel("Study Time Level (1 = Low, 4 = High)")
plt.ylabel("Average Final Grade")
plt.title("Average Final Grade by Study Time")

plt.grid(axis='y')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    avg.index,
    avg.values,
    marker='o',
    linestyle='-',
    color='purple'
)

plt.xlabel("Study Time Level")
plt.ylabel("Average Final Grade")
plt.title("Trend of Grades with Study Time")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    student_df["studytime"],
    student_df["G3"],
    color='teal',
    alpha=0.7
)

plt.xlabel("Study Time Level")
plt.ylabel("Final Grade")
plt.title("Student Study Time vs Final Grade")

plt.grid(True)

plt.show()
```

## Graph Analysis

The bar and line charts show that students who spend more time studying generally achieve higher grades.  
The scatter plot also shows a positive relationship between study time and performance.


## Interpretation of Results

The graph shows that students with higher study time levels generally have higher average final grades.

This indicates a positive relationship between study time and academic performance.
